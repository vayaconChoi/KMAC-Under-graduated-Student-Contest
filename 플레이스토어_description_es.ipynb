```python
from google_play_scraper import search, app
import pandas as pd

# Search term for Kiggle games
DEVELOPER_NAME = 'KIGLE'

def fetch_kiggle_games(developer_name):
    try:
        # Search for games by developer name in Spain
        games = search(
            developer_name,
            n_hits=50,  # Number of results to fetch per request
            lang='ja',  # Language set to Spanish
            country='us'  # Target the Spanish Play Store
        )
        if not games:
            print("No games found for this developer.")
        return games
    except Exception as e:
        print(f"Error occurred during search: {e}")
        return []

def fetch_game_details(app_id):
    try:
        # Fetch app details
        details = app(
            app_id,
            lang='ja',  # Language set to Spanish
            country='us'  # Target the Spanish Play Store
        )
        return details
    except Exception as e:
        print(f"Error occurred fetching details for {app_id}: {e}")
        return {}

def scrape_kiggle_games():
    # Step 1: Fetch all games associated with Kiggle in Spain
    games_list = fetch_kiggle_games(DEVELOPER_NAME)

    if not games_list:
        print("No games were returned from the search.")
        return None

    # Step 2: Get detailed info for each game
    detailed_games_info = []
    for game in games_list:
        app_id = game.get('appId')
        if app_id:
            details = fetch_game_details(app_id)
            detailed_games_info.append(details)

    # Step 3: Convert the result to a DataFrame for easy viewing/manipulation
    df = pd.DataFrame(detailed_games_info)

    # Step 4: Save the DataFrame to a CSV file or print it
    if not df.empty:
        df.to_csv('kiggle_games_info_es.csv', index=False)
        print("Data successfully saved to CSV.")
    else:
        print("No data to save.")

    return df

```

```python
result = scrape_kiggle_games()
result
```

```python
import re
import pandas as pd
import nltk
from nltk.corpus import stopwords

# NLTK 불용어 다운로드
nltk.download('stopwords')

# 영어 불용어 목록 가져오기
spanish_stopwords = set(stopwords.words('spanish'))

# 텍스트 전처리 및 불용어 제거 함수
def preprocess_spanish_text(text):
    # Step 1: 특수 문자 제거 및 공백 정리
    text_rmv = re.sub(r'[-=+,#/\?:^.@*\"※~ㆍ!』‘|\(\)\[\]`\'…》\”\“\’·]', ' ', text)
    text_rmv = re.sub(r'[\n\r]', ' ', text_rmv)  # 줄바꿈 제거
    text_rmv = re.sub(r'[^\w\s]', ' ', text_rmv)  # 특수 문자 제거
    text_rmv = re.sub(r'\s+', ' ', text_rmv)  # 다중 공백 제거

    # Step 2: 소문자 변환
    text_rmv = text_rmv.lower()

    # Step 3: 단어를 분리하여 불용어 제거
    tokens = text_rmv.split()
    filtered_words = [word for word in tokens if word not in spanish_stopwords]

    # Step 4: 불용어를 제거한 단어들로 다시 결합
    return ' '.join(filtered_words)

# DataFrame의 'description' 열에 전처리 적용 (영어 불용어 제거)
result['processed_description'] = result['description'].apply(preprocess_spanish_text)

# 전처리된 데이터 확인 (description과 processed_description 열만 출력)
print(result[['description', 'processed_description']].head())

# 결과를 CSV 파일로 저장
result.to_csv('processed_espanol_descriptions.csv', index=False)
```
