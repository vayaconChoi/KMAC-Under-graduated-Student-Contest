```python
from google_play_scraper import search, app
import pandas as pd

# Search term for Kiggle games
DEVELOPER_NAME = 'KIGLE'

def fetch_kiggle_games(developer_name):
    # Search for games by developer name
    games = search(
        developer_name,
        n_hits=50,  # Number of results to fetch per request
        lang='en',
        country='us'
    )
    return games

def fetch_game_details(app_id):
    return app(
        app_id,
        lang='en',
        country='us'
    )

def scrape_kiggle_games():
    # Step 1: Fetch all games associated with Kiggle
    games_list = fetch_kiggle_games(DEVELOPER_NAME)

    # Step 2: Get detailed info for each game
    detailed_games_info = []
    for game in games_list:
        app_id = game['appId']
        details = fetch_game_details(app_id)
        detailed_games_info.append(details)

    # Step 3: Convert the result to a DataFrame for easy viewing/manipulation
    df = pd.DataFrame(detailed_games_info)

    # Step 4: Save the DataFrame to a CSV file or print it
    df.to_csv('kiggle_games_info.csv', index=False)
    return df
```

```python
result = scrape_kiggle_games()
result.head()
```

```python
import re
import pandas as pd
from konlpy.tag import Komoran

# ë¶ˆìš©ì–´ íŒŒì¼ ê²½ë¡œ
stopwords_file_path = '/content/drive/MyDrive/á„á…©á„á…©á„‡á…µ_R&Dá„á…µá†·/á„‰á…¡á„‹á…­á†¼á„ƒá…¦á„‹á…µá„á…¥/stop_word_korean.txt'

# ë¶ˆìš©ì–´ íŒŒì¼ ì½ê¸°
with open(stopwords_file_path, 'r', encoding='utf-8') as file:
    stopwords = file.read().splitlines()

# Komoran í˜•íƒœì†Œ ë¶„ì„ê¸° ì´ˆê¸°í™”
komoran = Komoran()

# í…ìŠ¤íŠ¸ ì „ì²˜ë¦¬ ë° ëª…ì‚¬ ì¶”ì¶œ í•¨ìˆ˜
def preprocess_text(text):
    # Step 1: íŠ¹ìˆ˜ ë¬¸ì ì œê±° ë° ê³µë°± ì •ë¦¬
    text_rmv = re.sub(r'[-=+,#/\?:^.@*\"â€»~ã†!ã€â€˜|\(\)\[\]`\'â€¦ã€‹\â€\â€œ\â€™Â·]\r\n', ' ', text)
    text_rmv = re.sub(r'[\n\r]', ' ', text_rmv)  # ì¤„ë°”ê¿ˆ ì œê±°
    text_rmv = re.sub(r'[^\w\s]', ' ', text_rmv)  # íŠ¹ìˆ˜ ë¬¸ì ì œê±°
    text_rmv = re.sub(r'\s+', ' ', text_rmv)  # ë‹¤ì¤‘ ê³µë°± ì œê±°

    # Step 2: íŠ¹ì • íŒ¨í„´ êµì²´ ('ì½” ì½”ë¹„', 'ì½”ì½” ë¹„' ë“± ë‹¤ì–‘í•œ ë³€í˜•ì„ 'ì½”ì½”ë¹„'ë¡œ í†µì¼)
    text_rmv = re.sub(r'ì½”\s*ì½”\s*ë¹„', 'ì½”ì½”ë¹„', text_rmv)  # ëª¨ë“  í˜•íƒœì˜ 'ì½” ì½”ë¹„'ë¥¼ 'ì½”ì½”ë¹„'ë¡œ í•©ì¹¨

    # Step 3: ëª…ì‚¬ ì¶”ì¶œ (Komoran ì‚¬ìš©)
    tokens = komoran.nouns(text_rmv)

    # Step 4: ë¶ˆìš©ì–´ ì œê±°
    filtered_words = [word for word in tokens if word not in stopwords]

    # ìµœì¢… í‚¤ì›Œë“œ ë°˜í™˜ (ê³µë°±ìœ¼ë¡œ êµ¬ë¶„)
    return ' '.join(filtered_words)

# DataFrameì— ì „ì²˜ë¦¬ ì ìš©
result['keywords'] = result['description'].apply(preprocess_text)

# ì „ì²˜ë¦¬ëœ ë°ì´í„° í™•ì¸ (descriptionê³¼ keywords ì—´ë§Œ ì¶œë ¥)
print(result[['description', 'keywords']].head())

# ê²°ê³¼ë¥¼ CSV íŒŒì¼ë¡œ ì €ì¥
result.to_csv('processed_kiggle_games_info.csv', index=False)
```
