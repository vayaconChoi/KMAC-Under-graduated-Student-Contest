```python

# 불용어 파일 읽기
stopwords_file_path = '/content/drive/MyDrive/코코비_R&D팀/사용데이터/stop_word_korean.txt'
with open(stopwords_file_path, 'r', encoding='utf-8') as file:
    stopwords = file.read().splitlines()

# 형태소 분석기 초기화
okt = Okt()

# 텍스트 전처리 함수
def preprocess_text(text):
    # hanspell을 사용하여 띄어쓰기 보정
    checked = spell_checker.check(text)
    spaced_text = checked.checked

    # 특수문자 제거 및 다중 공백 제거
    spaced_text = re.sub(r'[^\w\s]', ' ', spaced_text)
    spaced_text = re.sub(r'\s+', ' ', spaced_text).strip()

    return spaced_text

# 모든 리뷰에 대한 명사 추출 및 불용어 제거
all_filtered_words = []

for content in cleaned_contents:
    # 텍스트 전처리
    processed_text = preprocess_text(content)

    # 명사 추출
    tokens = okt.nouns(processed_text)

    # 불용어 제거
    filtered_words = [word for word in tokens if word not in stopwords]

    # 결과를 리스트에 추가
    all_filtered_words.extend(filtered_words)

# 결과 출력
print(all_filtered_words)
```
