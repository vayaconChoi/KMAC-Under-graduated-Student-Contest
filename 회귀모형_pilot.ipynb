## 단계적 회귀법

```python
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# 후진 제거 함수 정의
def backward_elimination(X, y, significance_level=0.05):
    """
    후진 제거를 통해 변수를 선택하는 함수.
    p-value가 높은 변수부터 제거하며, 모든 변수의 p-value가 유의수준(significance_level) 이하일 때 멈춤.
    """
    X_const = sm.add_constant(X)
    while True:
        model = sm.OLS(y, X_const).fit()
        max_p_value = model.pvalues.max()
        if max_p_value > significance_level:
            excluded_feature = model.pvalues.idxmax()
            print(f"제거할 변수: {excluded_feature}, p-value: {max_p_value}")
            X_const = X_const.drop(columns=[excluded_feature])
        else:
            break
    return X_const, model

# 데이터 전처리 (위와 동일)
features = a
target = 'User acquisition'

# 독립 변수(X)와 종속 변수(y)를 분리
X = df[features]
y = df[target]

# 독립 변수와 종속 변수가 모두 숫자인지 확인하고, 필요시 숫자로 변환
X = X.apply(pd.to_numeric, errors='coerce')
y = pd.to_numeric(y, errors='coerce')

# 결측값 제거
X = X.dropna()
y = y[X.index]

# 시계열 데이터를 고려하여 트레인과 테스트로 나누기
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# 후진 제거를 통한 변수 선택
X_train_selected, final_model = backward_elimination(X_train, y_train)

# 최종 모델 요약 출력
print("\n최종 모델 요약:")
print(final_model.summary())

# 선택된 변수들로 테스트 세트 예측
X_test_const = sm.add_constant(X_test[X_train_selected.columns.drop('const')])  # 'const' 제외
y_pred = final_model.predict(X_test_const)

# 결과 비교
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print("\n실제값과 예측값 비교:")
print(comparison)

# 모델 성능 평가 (MSE와 R^2 스코어 출력)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"\nMean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

```python
# 필요한 라이브러리 임포트
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# target을 7일 뒤의 'User acquisition'으로 설정 (shift(-7))
df['User_acquisition_7days'] = df['User acquisition'].shift(-7)

# NaN이 있는 행 제거 (shift로 인해 마지막 7일은 예측 불가, 이를 제거)
df = df.dropna(subset=['User_acquisition_7days'])

# 시계열 데이터를 가미한 다중 회귀를 위해 필요한 피처 선택
# a 변수에 리스트 형태로 피처를 지정해 주세요. 예시: features = ['AppName', 'Country', 'CONTINENT', ...]
features = ['AppName', 'Country', 'CONTINENT', 'AD_LAN', 'Store_Visit', 'Store_Acquisition', 'Bound']

# 종속 변수는 7일 후의 'User acquisition'으로 설정
target = 'User_acquisition_7days'

# 독립 변수 (X)와 종속 변수 (y) 분리
X = df[features]
y = df[target]

# Min-Max Scaling 적용 (데이터프레임을 유지하며 스케일링)
scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=features)

# 시계열 데이터를 고려하여 트레인과 테스트로 나누기 (shuffle=False로 시계열 순서 유지)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, shuffle=False)

# 단계적 회귀법을 구현하는 함수
def stepwise_selection(X, y,
                       initial_list=[],
                       threshold_in=0.01,
                       threshold_out=0.05,
                       verbose=True):
    """ Perform a forward-backward feature selection
    based on p-value from statsmodels.api.OLS
    Arguments:
        X - pandas.DataFrame with candidate features
        y - list-like with the target
        initial_list - list of features to start with (column names of X)
        threshold_in - include a feature if its p-value < threshold_in
        threshold_out - exclude a feature if its p-value > threshold_out
        verbose - whether to print the sequence of inclusions and exclusions
    Returns:
        List of selected features
    """
    included = list(initial_list)
    while True:
        changed = False

        # forward step
        excluded = list(set(X.columns) - set(included))
        new_pval = pd.Series(index=excluded)
        for new_column in excluded:
            model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included + [new_column]]))).fit()
            new_pval[new_column] = model.pvalues[new_column]
        best_pval = new_pval.min()
        if best_pval < threshold_in:
            best_feature = new_pval.idxmin()
            included.append(best_feature)
            changed = True
            if verbose:
                print(f'Add  {best_feature:30} with p-value {best_pval:.6f}')

        # backward step
        model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included]))).fit()
        pvalues = model.pvalues.iloc[1:]  # excluding intercept
        worst_pval = pvalues.max()
        if worst_pval > threshold_out:
            worst_feature = pvalues.idxmax()
            included.remove(worst_feature)
            changed = True
            if verbose:
                print(f'Remove {worst_feature:30} with p-value {worst_pval:.6f}')

        if not changed:
            break

    return included

# 단계적 회귀법을 적용하여 최적의 변수 선택
result_features = stepwise_selection(X_train, y_train)

# 선택된 피처로 모델 재학습
X_train_selected = X_train[result_features]
X_test_selected = X_test[result_features]

# 다중 회귀 모델에 상수항 추가 (절편 추가)
X_train_const = sm.add_constant(X_train_selected)
X_test_const = sm.add_constant(X_test_selected)

# 다중회귀모델 적합
model = sm.OLS(y_train, X_train_const)
results = model.fit()

# 모델 요약 출력
print(results.summary())

# 예측값 계산
y_pred = results.predict(X_test_const)

# 결과 비교
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(comparison)

# 모델 성능 평가 (MSE와 R^2 스코어 출력)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```
